```python
import datetime
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import pandas_datareader as pdr
```

```python
import algtrd
```

```python
eurusd = pdr.get_data_yahoo('EURUSD=X', 
            start=datetime.datetime(2020, 1, 1), 
            end=datetime.datetime(2021, 8, 13))
```

```python
eurusd['diff'] = eurusd['Close'] - eurusd['Open']
data = eurusd.values
```

```python
result, min_balance, max_balance = algtrd.momentum(data, 1000, 50, 10, 5)
```

```python
print('result: {0: .3f}\nmin: {1: .3f}\nmax: {2: .3f}'.format(
            result, min_balance, max_balance))
    
const_profit = 0.0659 # USA 1 year bond
print('sharp: {0: .3f}\nprof/drdwn: {1: .3f}'.format(
        algtrd.sharp_koef(data, result, 1000, const_profit),
        algtrd.profit_drawdown_koef(result, 1000, min_balance)
))
```
